```python
import numpy as np
import matplotlib.pyplot as plt
import os

import torch as t
import torchvision as tv
from torchinfo import summary

from PIL import Image

import warnings
warnings.filterwarnings('ignore')
%matplotlib inline

#from google.colab import drive
#drive.mount('/content/drive')

device = t.device("cuda") if t.cuda.is_available() else t.device("cpu")
print(f'Using {device} for inference')
```

```python
ENV2 = tv.models.efficientnet_v2_s(weights = tv.models.EfficientNet_V2_S_Weights.DEFAULT)
# ENV2 = tv.models.efficientnet_b0(weights = tv.models.EfficientNet_B0_Weights.DEFAULT)
ENV2.eval();
```

```python
summary(ENV2, input_size = (1, 3, 224, 224), verbose = 1, col_names=("input_size", "output_size", "num_params"));
```

```python
def flatten(array):
    result = []
    for element in array:
        if hasattr(element, "__iter__"):
            result.extend(flatten(element))
        else:
            result.append(element)
    return result
layers = list(ENV2.children())
layers_flat = flatten(layers)
layers_flat
```
